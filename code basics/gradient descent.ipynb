```python
import pandas as pd
import numpy as np
```

```python
def grad_des(x,y):
    m_curr = 0
    b_curr = 0
    iterations = 1000
    learning_rate = 0.08
    n = len(x)
    for i in range(iterations):
        
        y_pred = m_curr*x + b_curr
        mse = 1/2*n*(sum(y-y_pred))
        md = -2/n*sum(x*(y-y_pred))
        bd = -2/n*sum(y-y_pred)
        m_curr = m_curr-learning_rate*md
        b_curr = b_curr-learning_rate*bd
        print("m {}, b {} ,i {} ,error {}".format(m_curr,b_curr,i,mse))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
grad_des(x,y)
```

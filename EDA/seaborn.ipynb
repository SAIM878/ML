```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x) 
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sin")
plt.title("sin wave")
plt.show()
```

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
languages = ['urdu','hindi','english','punjabi','pushto','hindko']
people = [50,5,20,20,2,1]
plt.bar(languages,people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```
